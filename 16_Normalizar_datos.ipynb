# Normalizar y reescalar datos

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Min-Max normalización

Toma el mínimo valor como cero y el máximo como uno

```python
from sklearn.preprocessing import MinMaxScaler
df = pd.DataFrame({'Edad': [25, 30, 35, 40], 'Salario': [50000, 60000, 85000, 100000]})
scaler = MinMaxScaler()

# Aplicar el reescalado
df_escalado = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)
print(df_escalado)
```

## Normalización estandarizada
Transforma los datos para que tengan una media de cero y una desviación estándar de uno

```python
from sklearn.preprocessing import StandardScaler
df = pd.DataFrame({'Edad': [25, 30, 35, 40], 'Salario': [50000, 60000, 85000, 100000]})
scaler = StandardScaler()

# Aplicar el reescalado
df_escalado = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)
print(df_escalado)
```

## Escalado robusto
Si tu conjunto de datos contiene muchos valores atípicos (outliers), tanto MinMaxScaler como StandardScaler pueden verse afectados. RobustScaler escala los datos utilizando la mediana y el rango intercuartílico, lo que lo hace resistente a valores extremos

```python
from sklearn.preprocessing import RobustScaler
df = pd.DataFrame({'Edad': [25, 30, 35, 40], 'Salario': [50000, 60000, 85000, 100000]})
scaler = RobustScaler()
df_robusto = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)
print(df_robusto)
```
