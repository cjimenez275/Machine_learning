Validación cruzada, consiste en validar que tan beno es un modelo de Machine learning tomando datos y dividiéndolos en k distintos conjuntos (unos 5 conjuntos). Se toman k-1 de ellos para entrenar el modelo y el último para validar, luego se repite esto interando y cambiando de tal manera que cada conjunto sirva para validar.
Ojo, validar es para ajustar hiperparámetros, debe dejarse algún conjunto externo de datos para probar.
Entrenar-Validar-Probar -> 60% - 20% - 20%

```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
```

```python
# En el archivo salario se refleja los datos de diferentes personas con edad
# años de estudio y un dato binario que dice si tiene ingresos altos o bajos
personas = pd.read_csv("salario.csv")
# En este caso vamos a probarlo para el modelo de ML de bosques aleatorios
bosque = RandomForestClassifier()

bosque.fit(personas[["edad", "estudio"]].values,
          personas["ingreso"].values)

# Realizamos la validación del modelo con los mismos datos que ha sido
# entrenado, por lo cual me dara un score muy bueno, pero no es realista
print(bosque.score(personas[["edad", "estudio"]].values,
                  personas["ingreso"].values))

# Realizamos la validación cruzada con cv = 5 pliegues, el resultado serán
# 5 valores de acuracy o score, uno por cada iteración,

print(cross_val_score(bosque,
                     personas[["edad", "estudio"]].values,
                     personas["ingreso"].values,
                     cv=5))


# podemos tomar el valor medio de ellos .mean() lo cual me dará un score mas bajo
# pero mas realista

print(cross_val_score(bosque,
                     personas[["edad", "estudio"]].values,
                     personas["ingreso"].values,
                     cv=5).mean())
```
